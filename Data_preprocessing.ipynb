```python
import pandas as pd 
df=pd.read_csv('life.csv')

df.head(15)
```

```python
import numpy as np
```

```python
df.isnull().sum() 
df.dropna(inplace=True) 
df.isnull().sum() 
df.shape
```

```python
counts = df['Country'].value_counts()
```

```python
rest = df[~df['Country'].isin(counts[counts < 5].index)]
```

```python
rest['Country'].value_counts()
```

```python
rest.shape
```

```python
data = rest.groupby('Country', group_keys = False).apply(lambda x: x.sample(3))
```

```python
data.head(15)
```

```python
data.shape
```

```python
data.describe()
```

```python
data['age_group'] = data['Life expectancy ']
data.head(15)
```

```python
data['age_group'] = pd.cut(data.age_group,bins=[35,50,65,80,100],labels=['35-50','50-65','65-80','80+'])
```
